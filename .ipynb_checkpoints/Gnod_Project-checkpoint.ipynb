```python
from bs4 import BeautifulSoup
import pandas as pd
import re
from pandas import json_normalize
```

```python
url = 'http://www.popvortex.com/music/charts/top-100-songs.php'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# print(soup.prettify())
```

```python
artists = []
titles = []

for title_artist_div in soup.select(".title-artist"):
    title = title_artist_div.find(class_="title")
    artist = title_artist_div.find(class_="artist")
    titles.append(title.get_text())
    artists.append(artist.get_text())
```

```python
print(artists)
```

```python
songs = pd.DataFrame({"title":titles,
                       "artist":artists
                      })
```

```python
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}
```

```python
tracks2['artist_dict'] = tracks2['track.album.artists'].apply(list_to_dict)
tracks2.head()
```

```python
tracks2.columns
```

```python
def expand_list_dict2(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index')
    df['song_id'] = row['track.id']
    return df

tracks2['artists_dfs'] = tracks2.apply(expand_list_dict2, axis=1)
tracks2['artists_dfs'][2]
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
```

```python
from sklearn.cluster import KMeans
```

```python
X = df_with_audiof[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)  
kmeans.fit(X_scaled)
```

```python
clusters = kmeans.predict(X_scaled)
pd.Series(clusters).value_counts().sort_index()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters,alpha=0.5)
plt.show()
```

```python
clusters
```

```python
import numpy as np
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled)
    silhouette.append(silhouette_score(X_scaled, kmeans.predict(X_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

```python
kmeans = KMeans(n_clusters=15, random_state=42)  
kmeans.fit(X_scaled)
```

```python
clusters = kmeans.predict(X_scaled)
pd.Series(clusters).value_counts().sort_index()
```

```python
def get_cluster_from_imput(x):
    search_results = sp.search(q=x, limit=1)
    song_id = search_results['tracks']['items'][0]['id']
    audio_features = sp.audio_features(song_id)
    audio_features_df = json_normalize(audio_features)
    X = audio_features_df[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
    # print('X', X)
    X_scaled = scaler.transform(X)
    # print('X_scaled', X_scaled)
    cluster = kmeans.predict(X_scaled)
    return cluster[0]
```

```python
import random

def pick_one_random_song(x):
    filtered_df = df_with_audiof[df_with_audiof['clusters'] == x]
    random_row = filtered_df.sample(n=1, random_state=random.seed())
    return random_row
```
