```python
from bs4 import BeautifulSoup
import pandas as pd
import re
from pandas import json_normalize
```

```python
url = 'http://www.popvortex.com/music/charts/top-100-songs.php'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# print(soup.prettify())
```

```python
artists = []
titles = []

for title_artist_div in soup.select(".title-artist"):
    title = title_artist_div.find(class_="title")
    artist = title_artist_div.find(class_="artist")
    titles.append(title.get_text())
    artists.append(artist.get_text())
```

```python
print(artists)
```

```python
songs = pd.DataFrame({"title":titles,
                       "artist":artists
                      })
```

```python
url2 = 'https://en.wikipedia.org/wiki/List_of_songs_recorded_by_Blink-182'
page2 = requests.get(url2)
```

```python
soup2 = BeautifulSoup(page2.content, 'html.parser')
```

```python
# soup2
```

```python
blink_songs = []
selected_elements_blink = soup2.select('table th[scope="row"]')

for song in selected_elements_blink:
    blink_songs.append(song.get_text())
    # print(song.get_text())
```

```python
pattern = r'"(.*?)"'

blink_only_titles = []
for item in blink_songs:
    match = re.search(pattern, item)
    if match:
        blink_only_titles.append(match.group(1))
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    count = 0
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        count += 1
        print(count)
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
all_tracks = get_playlist_tracks("5S8SJdl1BDc0ugpkEvFsIL")
len(all_tracks)
```

```python
tracks2 = json_normalize(all_tracks)
```

```python
tracks2['track.album.artists'][0][0]
```

```python
def list_to_dict(x):
    return {i: x[i] for i in range(len(x))}
```

```python
tracks2['artist_dict'] = tracks2['track.album.artists'].apply(list_to_dict)
tracks2.head()
```

```python
tracks2.columns
```

```python
def expand_list_dict2(row):
    df = pd.DataFrame.from_dict(row['artist_dict'], orient='index')
    df['song_id'] = row['track.id']
    return df

tracks2['artists_dfs'] = tracks2.apply(expand_list_dict2, axis=1)
tracks2['artists_dfs'][2]
```

```python
tracks2['artists_dfs'][25]['name']
```

```python
artist_df2 = pd.DataFrame(columns=['external_urls', 'href', 'id', 'name', 'type', 'uri'])
for val in tracks2['artists_dfs']:
    #display(val)
    artist_df2 = pd.concat([artist_df2, val], axis=0)
    
# artist_df2
```

```python
df_merged2 = pd.merge(left=tracks2,
                    right=artist_df2,
                    how='inner',
                    left_on='track.id',
                    right_on='song_id')
df_merged2.head()
```
